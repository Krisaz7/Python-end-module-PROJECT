```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dataset.csv')
df
```

```python
df.head()
df.info()
```

```python
df.isnull().sum()
```

```python
df['Height']=np.random.randint(150,180+1,len(df))
df
```

```python
df.fillna(value={'Salary': 0,'College':'Unspecified'})
```

```python
 #How many are there in each Team 
team_count = df['Team'].value_counts()
team_count
```

```python
total_count=sum(team_count)
total_count
```

```python
#percentage=((team_count/total_count)*100)
(((df["Team"].value_counts()/total_count)*100).round(2))
```

```python
#2.Segregate the employees w.r.t different positions.
emp=df.groupby('Position')

for i,group_data in emp:
    print("Employees in Position:",i)
    print(group_data)
```

```python
#Find from which age group most of the employees belong to.

# Create the histogram plot
plt.hist(df['Age'])

# Customize the plot
plt.xlabel('Age')
plt.ylabel('No of times')
plt.title('HISTOGRAM')

# Display the plot
plt.show()
```

```python
max_spend=df.groupby(['Team', 'Position'])['Salary'].sum()
max_salary=max_spend.idxmax()
max_salary
print("The team and position, whose spending in terms of salary is high is ",max_salary[0],',',max_salary[1])
```

```python
data=df.drop(['Name','Team', 'Position','College', 'Height'], axis='columns')
corr = data.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation between Age and Salary ")
plt.show()
```
